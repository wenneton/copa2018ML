```python
import pandas as pd
#from sklearn.preprocessing import train_test_split
```

```python
df = pd.read_csv("dataset_rank_fifa.csv", sep=',')
```

```python
df.head()
```

```python
df0 = df.loc[df['classification'] == 0]
df = df.loc[df['classification'] != 0]
```

```python
df0.sample(5)
```

```python
df.sample(5)
```

```python
df_balanced = pd.concat([df,df0.iloc[:16,:]])
df_balanced.sample(5)
```

```python
df_new = df_balanced.drop(['country', 'year'], axis=1)
df_new
```

```python
df_new.corr()
```

```python
df_new['classification'].value_counts()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.metrics import accuracy_score as acc

feature_cols = [x for x in df_new.columns if x not in 'classification']

sss = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)

train_idx, test_idx = next(sss.split(df_new[feature_cols], df_new['classification']))
                           
X_train = df_new.loc[train_idx, feature_cols]
y_train = df_new.loc[train_idx, 'classification']

X_test = df_new.loc[test_idx, feature_cols]
y_test = df_new.loc[test_idx, 'classification']
```

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)

y_pred = lr.predict(X_test)
acc(y_test,y_pred)
```

```python
results = X_test

results['true'] = y_test
results['pred'] = y_pred

results
```
